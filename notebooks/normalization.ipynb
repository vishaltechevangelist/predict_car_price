```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../data/used_cars.csv')
df.head()
```

```python
df['age_of_car'] = df['model_year'].max() - df['model_year']
df['milage_of_car'] = df.apply(lambda x: re.sub('[,mi.]', '', x['milage']), axis=1).astype(int)
df['price_of_car'] = df.apply(lambda x: re.sub(r'[$,]', '', x['price']), axis=1).astype(int)
df.head()
```

```python
mean_price_of_car = df['price_of_car'].mean()
std_price_of_car = df['price_of_car'].std()
df['normalized_price'] = df.apply(lambda x: (x['price_of_car'] - mean_price_of_car)/std_price_of_car, axis=1)
df['normalized_price']
```

```python
mean_age_of_car = df['age_of_car'].mean()
std_age_of_car = df['age_of_car'].std()
df['normalized_age'] = df.apply(lambda x: (x['age_of_car'] - mean_age_of_car)/std_age_of_car, axis=1)
df['normalized_age']
```

```python
mean_milage_of_car = df['milage_of_car'].mean()
std_milage_of_car = df['milage_of_car'].std()
df['normalized_milage'] = df.apply(lambda x: (x['milage_of_car'] - mean_milage_of_car)/std_milage_of_car, axis=1)
df['normalized_milage']
```

```python
df2 = df[['normalized_age', 'normalized_milage', 'normalized_price']]
df2.head()
```
