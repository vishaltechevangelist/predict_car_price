```python
df['price_no_dollar_no_comma'] = df.apply(lambda x: re.sub(r'[$,]', '', x['price']), axis=1)
df['price_no_dollar_no_comma'] = df['price_no_dollar_no_comma'].astype(int)
df.head()
```
